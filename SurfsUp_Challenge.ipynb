```python
# Dependencies
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use('seaborn')

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
#Base.classes.keys()


# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract



# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_results = session.query(Measurement.date,Measurement.tobs).\
    filter(extract('month',Measurement.date)==6)

print(june_results)
```

```python
# 2. Convert the June temperatures to a list.
june_results = session.query(Measurement.date,Measurement.tobs).\
    filter(extract('month',Measurement.date)==6).all()

print(june_results)
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_results, columns=['date', 'June_temps'])
june_df
```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_results = session.query(Measurement.date,Measurement.tobs).\
    filter(extract('month',Measurement.date)==12)

print(december_results)
```

```python
# 7. Convert the December temperatures to a list.
december_results = session.query(Measurement.date,Measurement.tobs).\
    filter(extract('month',Measurement.date)==12).all()

print(december_results)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_df = pd.DataFrame(december_results, columns=['date','December_temps'])
december_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_df.describe()
```

```python
# June and December Data Frames Combined  
combined_df = june_df.merge(december_df,left_index=True,right_index=True)
combined_df
```

```python
# June Temperature Histogram
june_df.plot.hist(bins=20,color='coral',edgecolor='black') 
plt.tight_layout()
plt.title('June Temperatures')
plt.savefig('JuneTemps.png')
```

```python
# December Temperature Histogram
december_df.plot.hist(bins=20, color='lightblue',edgecolor='black')
plt.tight_layout()
plt.title('December Temperatures')
plt.savefig('DecemberTemps.png')
```

```python
# Combined histogram
combined_df.plot.hist(bins=40,edgecolor='black')
plt.tight_layout()
plt.title('June December Combined Temperatures')
plt.savefig('June_December_Temps.png')
```

```python
# June and December combined Stats
june_stats = june_df.describe()
decemer_stats = december_df.describe()

# Join June and December stats
combined_stats_df = june_stats.merge(decemer_stats,left_index=True,right_index=True)
combined_stats_df
```

```python
# Comparison using Box and Whiskers Plot

# Plot the Box and Whiskers
x_labels = ["June Temps", "December Temps"]
temps = [june_df["June_temps"], december_df["December_temps"]]
fig, ax = plt.subplots(figsize=(10,6))
ax.boxplot(temps,labels=x_labels)

#Ticks and Labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)

ax.set_yticks(np.arange(55,95, step=2.0))
ax.set_title('June and December Temperatures', fontsize=18)
ax.set_ylabel('Temperature (F)',fontsize =14)
ax.set_xlabel("Month",fontsize =14)
```
